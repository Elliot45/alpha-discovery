```python
import pandas as pd
import matplotlib.pyplot as plt

# Chargement avec saut de la première ligne
df = pd.read_csv("../data/btc_2025_1min.csv", skiprows=1)

# Renommage et parsing
df.rename(columns={"Date": "timestamp", "Open": "price"}, inplace=True)
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp")

# Affichage
print(df[["timestamp", "price"]].head())

# Visualisation
df.set_index("timestamp")["price"].plot(figsize=(12, 4), title="BTCUSDT - Price (1min)")
plt.show()
```
